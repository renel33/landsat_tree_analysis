```python
import shutil
import os
import geopandas as gpd
```

```python
gdf = gpd.read_file("/mnt/sdc/tree_density_and_coverage/shapefiles/sahel_downloader/ethiopia/ethiopia_grid_split9_split4_filtered_2017.gpkg")
tiles = gdf["unique_id"].to_list()
tiles = [f"coords_{tile}_Landsat" for tile in tiles]
in_dir = "/mnt/sdd/senegal_trends"
out_dir = "/mnt/sdd/ethiopia_trends"
os.makedirs(out_dir, exist_ok=True)
```

```python
for tile in os.listdir(in_dir):
    if tile in tiles:   
        shutil.copytree(os.path.join(in_dir, tile), os.path.join(out_dir, tile))
```
