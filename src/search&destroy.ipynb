```python
import os
import glob
import tqdm
import rasterio
```

```python
dir = "/mnt/sdd/Sahel_Landsat89"
suffix = "background" # "background" " masked" " thresh"
dirs = os.listdir(dir)
len(dirs)
for file in tqdm.tqdm(dirs):
    files = glob.glob(f"{dir}/{file}/**/*{suffix}.tif", recursive=True)
    [os.remove(file) for file in files]
```

```python
def save_as_jp2(filename):
    with rasterio.open(filename) as src:
        rst = src.read()
        out_meta = src.meta.copy()
        out_meta.update({"driver": "JP2OpenJPEG"})
        with rasterio.open(filename.replace(".tif", ".jp2"), "w", **out_meta) as dst:
            dst.write(rst)
        
    os.remove(filename)
```

```python
dir = "/mnt/sdd/Sahel_Landsat"
suffix = "image" # "background" " masked" " thresh"
dirs = os.listdir(dir)

for file in tqdm.tqdm(dirs):
    
    files = glob.glob(f"{dir}/{file}/**/*{suffix}.tif", recursive=True)
    if len(files) == 0:
        continue
    
    [save_as_jp2(file) for file in files]
    #print(f"Finished converting files in {file}")
    
```
