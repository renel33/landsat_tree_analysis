```python
import rasterio
import numpy as np
import glob
import os
import tqdm
import time
```

```python
time.sleep(60*60*10)

landsat_prediction_dir = '/mnt/sdd/ls789gf'
for tile_dir in tqdm.tqdm(os.listdir(landsat_prediction_dir)):
    
    rel_diff_fps = sorted(glob.glob(f"{landsat_prediction_dir}/{tile_dir}/*.tif"))
    
    rel_diff  = [f for f in rel_diff_fps if f == f"{landsat_prediction_dir}/{tile_dir}/relative_diff_silver_sweep_9.tif"]
    
    if len(rel_diff) > 0:
        rel_diff = rel_diff[0]
        
        if os.path.exists(rel_diff.replace('relative_diff', 'significant_relative_diff')):
            continue
    else:
        continue
    
    trend = [f for f in rel_diff_fps if 'trend' in f]
    
    if len(trend) > 0:
        trend = trend[0]
    else:
        continue
    
    with rasterio.open(rel_diff) as src:
        rel_diff_arr = src.read()
        out_meta = src.meta.copy()
        
    with rasterio.open(trend) as src:
        trend_arr = src.read()
        # Perform calculations with error handling for division by zero
        
    with np.errstate(divide='ignore', invalid='ignore'):
        sigmask = np.where(trend_arr[1] <= 0.05, 1, 0)
    
    sig_rel_diff = rel_diff_arr[1] * sigmask
    out_array = np.concatenate([rel_diff_arr, sig_rel_diff[np.newaxis, :, :]], axis=0)
    out_meta.update(count=out_array.shape[0])
    out_fp = rel_diff.replace('relative_diff', 'significant_relative_diff')
    with rasterio.open(out_fp, 'w', **out_meta) as dst:
        dst.write(out_array)
    
    print(f"Saved {out_fp}")
```
